```python
#Q1

#Create an array using numPy and perform the following operations on it:
#Create a one dimensional array (vector) iterate through all the elements and find the mean of the vector.

import numpy as np
arr=[1,2,3,4,5,6,8]
narr=np.array(arr)
sum=0
for i in narr:
    sum+=i
print("mean is", sum/len(narr))
print(narr.mean())
```

```python
Create a 3X3 array and find the sum of all elements in the second column
#Create array using other standard methods like ones and empty by taking the zeros as reference


arr = np.array([[[1, 2, 3], [4, 5, 6]], [[13, 14, 15], [16, 17, 18]],[[7, 8, 9], [10, 11, 12]]])
print(arr)
print(arr[1].sum(),"is sum of all elements in second column")
a=np.eye(3)
b=np.zeros((3,3))
print(a)
print(b)
print(a[1].sum())
print(b[1].sum())
```

```python
#Create a matrix and perform the below operations on it
#Get the diagonal elements and calculate the sum of all the diagonal elements.
#Calculate the column wise and row wise mean of the created matrix
x = np.matrix([[1,2,3],[4,5,6],[7,8,9]])
print(x)
print(np.diagonal(x).sum(),"sum of diagonal elements")
print(x.mean(axis=1),"row wise mean")
print(x.mean(axis=0),"column wise mean")
```

```python
#Q2

import re

def display():
    choice=int(input("Welcome to the Device Management System\n enter your choice. choose from below option\n1. view all devices\n2. add a device\n3. delete a device\n4.update a device\n5.exit the program : "))
    return choice

def view_file():
    with open(r"C:\Users\Yogi\Desktop\device.txt", 'r+') as rf:
        l=rf.readlines()
        clean_list=[]
        for i in l:
            clean_list.append(re.sub("\n","",i))
    str=""
    for i in clean_list:
        str+=i+"\n"
    print(str)
    return clean_list


def device_add():
    user_input=input("enter device name you want to add")
    with open(r"C:\Users\Yogi\Desktop\device.txt", 'a') as fr:
        fr.write("\n")
        fr.write(user_input)
    print(user_input, "is added")
    return None



def delete_device():
    user_del=input("enter device name you want to delete: ")
    clean_list=view_file()
    print(clean_list)
    for i in clean_list:
        if user_del==i:
            print("removing device: ", i)
            clean_list.remove(i)
            print(user_del," user removed")
    
    with open(r"C:\Users\Yogi\Desktop\device.txt", 'w+') as fr:
        for x in clean_list:
            fr.write(x)
            fr.write("\n")
    return clean_list

def update_device():
    user_input=input("enter device name you want to update: ")
    clean_list=view_file()
    user_up=input("enter new device name")
    print(clean_list)
    for i in clean_list:
        if user_input==i:
            clean_list.remove(i)
            clean_list.append(user_up)
            print(user_up, "is updated")
    with open(r"C:\Users\Yogi\Desktop\device.txt", 'w+') as fr:
        for x in clean_list:
            fr.write(x)
            fr.write("\n")
    return clean_list


def main():
    var=display()
    if var==1:
        view_file()
        if input("Do you want to continue?(y/n)")=="y":
            main()
        else:
            print("bye for now")
            
    elif var==2:
        device_add()
        if input("Do you want to continue?(y/n)")=="y":
            main()
        else:
            print("bye for now")
    
    elif var==3:
        delete_device()
        if input("Do you want to continue?(y/n)")=="y":
            main()
        else:
            print("bye for now")
        
    elif var==4:
        update_device()
        if input("Do you want to continue?(y/n)")=="y":
            main()
        else:
            print("bye for now")
    
    elif var==5:
        print("Bye! Thanks for using the application !")
        
    
    else: 
        print("invalid option: Try Again")
        if input("Do you want to continue?(y/n)")=="y":
            main()
        else:
            print("bye for now")
            
if __name__=="__main__":
    main()
```
