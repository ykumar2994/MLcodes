# Image processing with OpenCV: Load image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.png")
```

```python
print(type(img))
```

# Image processing with OpenCV: Dispaly image

```python
plt.imshow(img)
```

# Image processing with OpenCV: Convert Color space

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# Image processing with OpenCV: Image resizing

```python
scale = 20
new_width = int(img_rgb.shape[1]*scale/100)
new_height = int(img_rgb.shape[0]*scale/100)
dim = (new_width, new_height)

#resize image
resized = cv2.resize(img_rgb, dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized)
print(resized.shape)
```

# Image processing with OpenCV: Crop Images

```python
img_crop = img_rgb[80:280, 50:288]
plt.imshow(img_crop)
print(img_crop.shape)
```

# Image processing with OpenCV: Save image

```python
cv2.imwrite('pic_2.png', img)
cv2.imwrite('pic_3.png', resized)
cv2.imwrite('pic_4.png', img_crop)
```

# Image processing with Scikit-image: Load image

```python
import numpy as np
import skimage
from skimage import io
```

```python
img = io.imread("image.png")
```

```python
print(type(img))
print(img.shape)
```

# Image processing with Scikit-image: Dispaly image

```python
plt.imshow(img)
```

# Image processing with Scikit-image: Image resizing

```python
from skimage.transform import resize

#resize image
resized = resize(img, (400,400))
plt.imshow(resized)
print(resized.shape)
```

# Image processing with Scikit-image: Crop Images

```python
img_crop = img[80:280, 50:288]
plt.imshow(img_crop)
print(img_crop.shape)
```

# Image processing with Scikit-image: Save image

```python
io.imsave('pic_2b.png', img)
io.imsave('pic_3b.png', resized)
io.imsave('pic_4b.png', img_crop)
```

# Image processing with Pillow: Load image

```python
import numpy as np
from PIL import Image
```

```python
img = Image.open("image.png")
```

```python
print(type(img))
print(img.size)
```

# Image processing with Pillow: Dispaly image

```python
plt.imshow(img)
```

# Image processing with Pillow: Image resizing

```python
size = (50,50)

#resize image
resized = img.resize(size)
plt.imshow(resized)
print(resized.size)
```

# Image processing with Pillow: Crop Images

```python
img_crop = img.crop((80,50,280,288))
plt.imshow(img_crop)
print(img_crop.size)
```

# Image processing with Pillow: Save image

```python
img.save('pic_2c.png')
resized.save('pic_3c.png')
img_crop.save('pic_4c.png')
```
