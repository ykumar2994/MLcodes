# Data Insight - Covid data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

covid = pd.read_csv('ongoing_outbreaks.csv')
```

```python
covid.describe()
#### max cases in a day =426
#### min cases in a day = 1
```

```python
covid.sort_values(by='number_ongoing_outbreaks').head(10)
### highest number of cases seen in '1 Congregate Care' outbreak_group 
```

```python
covid.groupby('date').sum().sort_values(by='number_ongoing_outbreaks',ascending=False).head(10)
#highest number of cases seen in month of jan 22
```

```python
covid.plot(kind='line',x='date',y='number_ongoing_outbreaks')
#line plot for covid cases
```

```python
covid_group= covid.groupby('outbreak_group').sum()
print(covid_group)
```

```python
#total no of cases in outbreak_group
covid_group.plot(kind='barh',title='total cases in outbreak_group')
```

```python
# scatter plot 
covid.plot(kind='scatter',x='outbreak_group',y='number_ongoing_outbreaks')
```

```python
subgroup_dataset=covid.groupby('outbreak_subgroup').sum()
```

```python
subgroup_dataset.plot(kind='bar',title='total cases in outbreak_subgroup')
#total cases in outbreak_subgroup. highset cases in '1 Long-Term Care Homes' outbreak_subgroup.
```
